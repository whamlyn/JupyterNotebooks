```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
welldata = pd.read_csv(r'./data/basindb_wells_grandbanks.csv', skiprows=3)
locdata = pd.read_csv('./data/basindb_wells_grandbanks_loc.csv', skiprows=3)
```

```python
fig = plt.figure(num=1, figsize=(10,10)); fig.clf()
ax1 = fig.add_subplot(111)
ax1.plot(locdata['Longitude (NAD83)'], locdata['Latitude (NAD83)'], '.')
for well, lat, lon in zip(locdata['Well Name'], locdata['Latitude (NAD83)'], locdata['Longitude (NAD83)']):
    ax1.text(lon, lat, well, ha='left', va='bottom', fontsize=7)
```
